```python
import pandas as pd

# Load your parquet file
df = pd.read_parquet(r"C:\Users\zulaika.sukarno\Desktop\2025\PI Calculator\pi_calculator_project\Avg Price PI Aug25 (location).parquet")

# Clean and prepare data
df = df.dropna(subset=['subsubcategory', 'location', 'unit_raw', 'avg_price'])
df = df[df['subsubcategory'].astype(str).str.strip() != '']

# Group by location, unit, and item name
grouped_df = df.groupby(['location', 'unit_raw', 'subsubcategory'])['avg_price'].mean().reset_index()

# Rename columns to match our structure
grouped_df.columns = ['location', 'unit', 'name', 'price']
grouped_df['price'] = grouped_df['price'].round(2)

# Export to TypeScript format
items = grouped_df.to_dict('records')

# Generate TypeScript code
with open(r'C:\Users\zulaika.sukarno\Desktop\itemData.ts', 'w') as f:
    f.write('export const itemData = ')
    f.write(str(items).replace("'", '"'))
    f.write(';')

print(f"Exported {len(items)} items!")
```
